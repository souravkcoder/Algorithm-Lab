```python
def extract_answer(qa_chain_output):
    lines = qa_chain_output.split('\n')
    for line in lines:
        if line.startswith('Answer:'):
            return line.split(':', 1)[1].strip()
    return None


question = "মুক্তিযুদ্ধে কয়টি সেক্টর ছিল?"
result = qa_chain.invoke(question)
print(result)
```
